```python
from fastapi import FastAPI, File, UploadFile, HTTPException
from fastapi.encoders import jsonable_encoder
from transformers import pipeline
from fastapi.middleware.cors import CORSMiddleware
import torch
from PIL import Image
from io import BytesIO
from pyngrok import ngrok
import nest_asyncio
import base64
import logging
# Initialize FastAPI app
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
app = FastAPI()
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Allows all origins (for development only)
    allow_credentials=True,
    allow_methods=["*"],  # Allows all methods (GET, POST, etc.)
    allow_headers=["*"],  # Allows all headers
)
```

```python

pipe = pipeline(
    "image-text-to-text",
    model="google/gemma-3-4b-it",
    device="cuda",
    torch_dtype=torch.bfloat16
)
```

```python

MEDICAL_SYSTEM_PROMPT = """You are a certified pharmaceutical analysis system. Your responses MUST:
1. Identify active ingredients and their concentrations
2. List excipients/additives
3. Highlight potential allergens (lactose, gluten, etc.)
4. Provide generic alternatives
5. Flag any controlled substances
6. Use medical terminology from WHO guidelines"""

@app.get("/test")
async def test():
    return {"message": "CORS test successful!"}
@app.post("/analyze_medicine/")
async def analyze_medicine(image: UploadFile = File(...)):
    try:
        image_data = await image.read()
        image_pil = Image.open(BytesIO(image_data))

        # Convert image to base64
        buffered = BytesIO()
        image_pil.save(buffered, format="PNG")
        image_base64 = base64.b64encode(buffered.getvalue()).decode("utf-8")

        # Prepare messages
        messages = [
            {
                "role": "system",
                "content": [{"type": "text", "text": MEDICAL_SYSTEM_PROMPT}]
            },
            {
                "role": "user",
                "content": [
                    {"type": "image", "image": f"data:image/png;base64,{image_base64}"},
                    {"type": "text", "text": "What are the contents of this medicine? Can you suggest alternative tablets?"}
                ]
            }
        ]

        # Run the model
        output = pipe(text=messages, max_new_tokens=200)

        # Log the output for debugging
        logging.info(f"Model Output: {output}")

        # Ensure output is JSON serializable
        if isinstance(output, dict):
            response = output
        elif isinstance(output, list):
            response = {"analysis": [str(item) for item in output]}
        else:
            response = {"analysis": str(output)}

        return jsonable_encoder(response)

    except Exception as e:
        logging.error(f"Error in /analyze_medicine/: {e}", exc_info=True)
        raise HTTPException(status_code=500, detail=str(e))
# Allow multiple threads to run the FastAPI app
```
